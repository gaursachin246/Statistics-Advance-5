```python
### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

from scipy.stats import norm

# Define the sample mean and standard deviation
sample_mean = 50
sample_std_dev = 5

# Define the sample size (assuming a large sample size for simplicity)
sample_size = 1000

# Calculate the standard error
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the margin of error (critical value * standard error)
margin_of_error = norm.ppf(0.975) * standard_error

# Calculate the 95% confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"The 95% confidence interval is ({lower_bound}, {upper_bound})")
```

```python
### Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.
#To conduct a chi-square goodness of fit test to determine if the distribution of M&M colors in a bag matches the expected distribution, follow these steps:

#Collect observed data: Gather the actual counts of each color from the M&Ms in the bag.

#Define the expected proportions: According to the problem, the expected proportions are:

#Blue: 20%
#Orange: 20%
#Green: 20%
#Yellow: 10%
#Red: 10%
#Brown: 20%
#Calculate the expected counts: Based on the total number of M&Ms in the bag and the expected proportions.

#Perform the chi-square test: Compare the observed counts to the expected counts using the chi-square test.

#Hereâ€™s how you can perform this test using Python:

import scipy.stats as stats

# Observed counts of M&M colors (example data; replace with actual data)
observed_counts = [25, 30, 20, 10, 15, 30]  # Example data; replace with actual counts

# Expected proportions
expected_proportions = [0.20, 0.20, 0.20, 0.10, 0.10, 0.20]

# Total number of M&Ms
total_mms = sum(observed_counts)

# Calculate the expected counts based on the proportions
expected_counts = [p * total_mms for p in expected_proportions]

# Perform the chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed_counts, f_exp=expected_counts)

# Significance level
alpha = 0.05

# Print the results
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Determine if we reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution significantly differs from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to say the observed distribution differs from the expected distribution.")



```

```python
### Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in thepopulation who smoke.

from scipy.stats import norm

# Define the sample size and number of smokers
sample_size = 500
num_smokers = 60

# Calculate the sample proportion
sample_proportion = num_smokers / sample_size

# Calculate the standard error
standard_error = (sample_proportion * (1 - sample_proportion) / sample_size) ** 0.5

# Calculate the margin of error (critical value * standard error)
margin_of_error = norm.ppf(0.975) * standard_error

# Calculate the 95% confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print(f"The 95% confidence interval for the true proportion of smokers is ({lower_bound:.4f}, {upper_bound:.4f})")

```

```python
##Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

from scipy.stats import norm

# Define the sample mean and standard deviation
sample_mean = 75
sample_std_dev = 12

# Define the sample size (assuming a large sample size for simplicity)
sample_size = 1000

# Calculate the standard error
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the margin of error (critical value * standard error)
margin_of_error = norm.ppf(0.95) * standard_error

# Calculate the 90% confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"The 90% confidence interval is ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Define the degrees of freedom
df = 10

# Generate x values
x = np.linspace(0, 30, 100)

# Calculate the corresponding chi-square distribution values
y = chi2.pdf(x, df)

# Create the plot
plt.plot(x, y)

# Label the axes
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')

# Shade the area corresponding to a chi-square statistic of 15
plt.fill_between(x, y, where=(x >= 15), color='gray', alpha=0.3)

# Show the plot
plt.show()
```
